# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
imdb=requests.get('https://www.imdb.com/chart/top/',headers={'User_agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'})
imdb
```

```python
imdbsoup=BeautifulSoup(imdb.content)
imdbsoup
```

```python
imdbsoup.title.text
```

```python
# To extract movie title
name=[]
for title in imdbsoup.find_all('div',class_='ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b0691f29-9 klOwFB cli-title'):
    name.append(title.text.split('. ')[1])
name[:5]
```

```python
# To extract rating
rating=[]
for rate in imdbsoup.find_all('div',class_='sc-e2dbc1a3-0 ajrIH sc-b0691f29-2 bhhtyj cli-ratings-container'):
    a=rate.text
    b=a[:3]
    rating.append(b)
rating[:5]
```

```python
# To extract year of release
year=[]
for yr in imdbsoup.find_all('div',class_='sc-b0691f29-7 hrgukm cli-title-metadata'):
    a=yr.text
    b=a[:4]
    year.append(b)
year[:5]
```

```python
# Create dataframe with the extracted data
import pandas as pd
names=pd.DataFrame(name,columns=['Movie title'])
ratings=pd.DataFrame(rating,columns=['Rating'])
years=pd.DataFrame(year,columns=['Release year'])
```

```python
merged=pd.concat([names,ratings,years],axis=1)
merged 
```

```python
# This new dataframe 'imdb_data' has IMDB’s Top rated 100 movies data 
imdb_data=merged.head(100)
imdb_data
```

# 3) Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL.

```python
dine=requests.get('https://www.dineout.co.in/hyderabad-restaurants/welcome-back')
dine
```

```python
dinesoup=BeautifulSoup(dine.content)
print(dinesoup.prettify())
```

```python
# (i) To extract restaurant name
name=[]
for res in dinesoup.find_all('a',class_='restnt-name ellipsis'):
    name.append(res.text)
names=pd.DataFrame(name,columns=['Restaurants'])
names
```

```python
# (ii) To extract cuisine
cuisine=[]
for cui in dinesoup.find_all('span', class_='double-line-ellipsis'):
    cuisine.append(cui.text.split('|')[-1])
cuisines=pd.DataFrame(cuisine,columns=['Cuisines'])
cuisines
```

```python
# (iii) To extract restaurant location
location=[]
for loc in dinesoup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(loc.text)
locations=pd.DataFrame(location,columns=['Location'])
locations
```

```python
# (iv) To extract restaurant ratings
rating=[]
for rate in dinesoup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(rate.text)
ratings=pd.DataFrame(rating,columns=['Rating'])
ratings
```

```python
# (v) To extract image url
img=[]
for im in dinesoup.find_all('img',class_='no-img'):
    img.append(im['data-src'])
imgs=pd.DataFrame(img,columns=['Image url'])
imgs.head()
```

```python
dineout=pd.concat([names,cuisines,locations,ratings,imgs],axis=1)
dineout.head()
```

# 4) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
pre=requests.get('https://presidentofindia.nic.in/index.php/former-presidents')
pre
```

```python
presoup=BeautifulSoup(pre.content)
presoup
```

```python
import pandas as pd
import regex as re
```

```python
name=[]
for i in presoup.find_all('h3'):
    name.append(i.text)
names=pd.DataFrame(name,columns=['President name'])
names
```

```python
rank=[]
for i in presoup.find_all('h5'):
    a=i.text
    b=re.findall(r'\d+',a)
    rank.append(b)
ranks=pd.DataFrame(rank,columns=['Rank'])
ranks
```

```python
presidents=pd.concat([names,ranks],axis=1)
presidents
```
